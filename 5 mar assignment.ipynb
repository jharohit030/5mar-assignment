```python
#Question 1:

from bokeh.plotting import figure, show
import numpy as np
import bokeh.io
bokeh.io.output_notebook()

x = np.random.rand(100)
y = np.random.rand(100)

plot = figure(title="Random Scatter Plot", x_axis_label="X", y_axis_label="Y")

plot.circle(x, y, size=10, color="blue", alpha=0.5)

show(plot)
```

```python
#Question 2:

In Bokeh, glyphs are geometric shapes that can be used to visualize data on a plot. Glyphs can represent points, lines, rectangles, circles, and other shapes, and they can be customized with various attributes like color, size, alpha (transparency), and more

from bokeh.plotting import figure, show
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 50, size=100)
plot = figure(title="Random Scatter Plot", x_axis_label="X", y_axis_label="Y")
plot.circle(x, y, size=sizes, color="blue", alpha=0.5)
show(plot)
```

```python
#Question 3:

Here are some of the most common ways to customize the appearance of a Bokeh plot:

1.Set the plot title: we can set the title of the plot using the title attribute of the figure object.
    plot = figure(title="My Plot")
    
2.Set the axis labels: we can set the labels for the x- and y-axes using the x_axis_label and y_axis_label attributes of the figure object.
    plot = figure(x_axis_label="X Axis", y_axis_label="Y Axis")
    
3.Customize the legend: You can customize the appearance of the legend, including its position, font size, and style, using the legend attribute of the figure object.

    plot = figure()
    plot.legend.location = "top_left"
    plot.legend.label_text_font_size = "14pt"
    plot.legend.label_text_font_style = "bold"
    
4.Set the plot size: You can set the size of the plot using the plot_width and plot_height attributes of the figure object.
    plot = figure(plot_width=800, plot_height=400)
```

```python
#Question 4:

A Bokeh server is a Python program that allows you to create interactive plots and applications that can be updated in real time. With a Bokeh server, you can create plots that respond to user input, update data in real time, and interact with other Python libraries and data sources.

To use a Bokeh server, we need to create a Bokeh application, which is a Python function that defines the layout and behavior of the plot or application. The function takes a Document object as an argument, which represents the HTML document that will be displayed in the web browser.

from bokeh.models import ColumnDataSource
from bokeh.io import curdoc
import random

source = ColumnDataSource(data=dict(x=[random.random() for i in range(100)], y=[random.random() for i in range(100)]))
plot = figure()
plot.circle(x='x', y='y', source=source)

def update():
    new_data = dict(x=[random.random() for i in range(100)], y=[random.random() for i in range(100)])
    source.data = new_data


curdoc().add_root(plot)
curdoc().add_periodic_callback(update, 1000)
```

```python
#Question 5:

To embed a Bokeh plot into a web page or dashboard using Flask or Django, we can use the components() function from the bokeh.embed module, which generates the HTML and JavaScript code needed to display a Bokeh plot in a web page.
Here example of flask,

from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # create a Bokeh plot
    plot = figure()
    plot.circle([1, 2, 3], [4, 5, 6])

    # generate the HTML and JavaScript code for the plot
    script, div = components(plot)

    # render the template and pass the plot code to it
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
```
